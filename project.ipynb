```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("google_books_dataset.csv")
df.head(5)          # первые строки
```

```python
df.tail(5)          # последние строки
```

```python
df.info()          # типы данных и пропуски
```

```python
print(df.isna().sum())                     # сколько пропусков в каждом столбце
```

```python
for i in df:
  df = df.dropna(subset=[i])    # убрать пустые строки
print(df.isna().sum())          # сколько пропусков в каждом столбце
```

```python
print(df.duplicated().sum())
```

```python
print(df.describe())
```

```python
print(df["authors"].unique())      # уникальные авторы
```

```python
print(df["categories"].unique())      # уникальные категории
```

```python
numerical_cols_for_corr = [col for col in df.select_dtypes(include=np.number).columns]

plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols_for_corr].corr(), annot=True, cmap='coolwarm')
plt.title('Корреляция численных столбцов')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['page_count'])
plt.title('Распределение количества страниц')
plt.xlabel('Число страниц')
plt.ylabel('Частота')
plt.show()
```

```python
top_10_authors = df['authors'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_authors.values, y=top_10_authors.index)
plt.title('Топ 10 авторов по числу книг')
plt.xlabel('Число книг')
plt.ylabel('Автор')
plt.show()
```

```python
df['published_year'] = df['published_date'].apply(lambda x: x.split('-')[0])

books_per_year = df.groupby('published_year').size().sort_index()

plt.figure(figsize=(15, 7))
sns.lineplot(x=books_per_year.index, y=books_per_year.values)
plt.title('Число опубликованных книг в год')
plt.xlabel('Год публикации')
plt.ylabel('Число книг')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['page_count'], y=df['list_price'])
plt.xlabel('Число страниц')
plt.ylabel('Цена книги')
plt.show()
```

```python
average_ratings_per_author = df.groupby('authors')['average_rating'].mean()
print(average_ratings_per_author)
```

```python
max_pages_per_author = df.groupby('authors')['page_count'].max()
print(max_pages_per_author.head(10))
```
